```python
marketplandf = data[[0,1]]
```

```python
marketplandf=marketplandf.drop([0])
```

```python
marketplandf.groupby(0)[1].apply(list)
```

```python
marketplandf.groupby([1]).get_group('WIRELESS')
```

```python
marketplandf.rename(columns={0: 'city', 1: 'data request type'}, inplace=True)
marketplandf.head(2)
```

```python
marketplandf=pd.DataFrame({'count' : marketplandf.groupby( [ "data request type","city"]).size()}).reset_index()
marketplandf
```

```python
marketplandf_DTV=marketplandf.loc[marketplandf["data request type"] =="DTV"]
marketplandf_DTV
```

```python
import matplotlib.pyplot as plt
colors = ["yellow", "green", "blue", "red"]
plt.pie(
    # using data total)arrests
    marketplandf_DTV['count'],
    # with the labels being officer names
    labels=marketplandf_DTV['city'],
    # with no shadows
    shadow=False,
    # with colors
    colors=colors,
    # with one slide exploded out
    explode=(0, 0, 0.15, 0),
    # with the start angle at 90%
    startangle=90,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
    )

# View the plot drop above
plt.axis('equal')

# View the plot
plt.tight_layout()
plt.title(" AT&T DTV network distribution across cities",fontsize=30)
plt.show()
```

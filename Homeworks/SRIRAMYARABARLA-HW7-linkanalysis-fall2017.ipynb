```python
# %load nfl.py
import scrapy

class nflItem(scrapy.Item):
    title = scrapy.Field()
    link = scrapy.Field()
    resp = scrapy.Field()
    
class nflSpider(scrapy.Spider):
    name = 'nfl'
    allowed_domains = ['nfl.com']
    start_urls = ['http://www.nfl.com/scores', 'http://www.nfl.com/fantasyfootball?icampaign=fty-nav-hp','http://www.nfl.com/news','https://www.nfl.com/standings']

    def parse(self, response):
        hxs = scrapy.Selector(response)
        titles = hxs.xpath('//ul/li')
        items = []
        for title in titles:
            item = nflItem()
            item["title"] = title.xpath("a/text()").extract()
            item["link"] = title.xpath("a/@href").extract()
            item["resp"] = response
            if item["title"] != []:
                items.append(item)

        return items
```

```python
import pandas as pd
p = []
data = pd.read_csv("items.csv")
data['link'] = 'http://www.nfl.com'+data['link']
r = data['resp'][0]
p.append((r.split()[-1]).split('>')[-2])
for i in range(len(data)):
    if data['resp'][i] == r:
        p.append(data['link'][i])
    else:
        r = data['resp'][i]
       
```

```python
nfl_items = list(pd.DataFrame(p)[0].unique())
```

```python
links = []
length = len(p)
for i, val in enumerate(p):
    if i < length-1:
        links.append((p[i], p[i+1]))
```

```python
import numpy as np
n = pd.DataFrame(index=nfl_items, columns=nfl_items)
m = n.replace(np.NaN, 0)
for i in links:
    m.loc[i] = 1.0
ar = np.array(m)
v = ar.sum(axis=1)
res = ar/v[:, np.newaxis]
st_mat = np.nan_to_num(res)
```

```python
from numpy import *

def pagerank(H):
    n= len(H)
    w = zeros(n)
    rho = 1./n * ones(n);
    for i in range(n):
        if multiply.reduce(H[i]== zeros(n)):
            w[i] = 1
    newH = H + outer((1./n * w),ones(n))
 
    theta=0.85
    G = (theta * newH) + ((1-theta) * outer(1./n * ones(n), ones(n)))
    #print rho
    for j in range(10):
        rho = dot(rho,G)
        #print rho
    
    return list(rho)
```

```python
r = pagerank(st_mat)
sort_val = ((pd.DataFrame(r)).sort_values(0, ascending=False)).head(10)
sort_index = list(sort_val.index)
```

```python
def hits(A):
    n= len(A)
    Au= dot(transpose(A),A)
    Hu = dot(A,transpose(A))
    a = ones(n); h = ones(n)
    #print a,h
    for j in range(5):
        a = dot(a,Au)
        a= a/sum(a)
        h = dot(h,Hu)
        h = h/ sum(h)
    return h
```

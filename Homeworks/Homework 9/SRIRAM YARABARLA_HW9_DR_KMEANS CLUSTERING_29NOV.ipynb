```python
data = pd.read_table("pollution2.csv", sep=',', header=None)
data.index.names = ['location']
data.columns.names = ['pollution']
data.head()
```

```python
import numpy as np
np_data = data.as_matrix()
np_data[0] 
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
a =le.fit_transform(data[0])
data[0]=a
data[0]
```

```python
data.loc[[0],0:1] = 1,2
data.loc[[0],2:3] = 3,4
data.loc[[0],4:5] = 5,6
data.loc[[0],6:7] = 7,8
data.loc[[0],8:9] = 9,10
data.loc[[0],10:11] = 11,12
data.loc[[0],12:13] = 13,14
data.head()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing   
pca = PCA(n_components=2)
pca.fit(data)
```

```python
 PCA(copy=True, n_components=2, whiten=False)
    
```

```python
data_2d = pca.transform(data)
```

```python
data_2d = pd.DataFrame(data_2d)
data_2d.index = data.index
data_2d.columns = ['PC1','PC2']
data_2d.head()
```

# K-means

```python
#All the following code is from the below citation[1]
from sklearn.cluster import KMeans
    
kmeans = KMeans(n_clusters=5)
clusters = kmeans.fit(data)
```

```python
data_2d['cluster'] = pd.Series(clusters.labels_, index=data_2d.index)
```

```python
%matplotlib inline
import numpy as np
    
data_2d.plot(
        kind='scatter',
        x='PC2',y='PC1',
        c=data_2d.cluster.astype(np.float), 
        figsize=(16,8))
```
